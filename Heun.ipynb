```python
import matplotlib.pyplot as plt
import numpy as np

def diferencial(x,y):
    return -2*x**3 +12*x**2 - 20*x + 8.5

def solucionexacta(x):
    return -(x**4)/2 + 4*x**3-10*x**2+8.5*x+1

def Heun(xi,xf,yi,h):
    n=((xf-xi)/h)+1
    x=np.empty(int(n))
    y=np.empty(int(n))
    y0=0
    x[0]=xi
    y[0]=yi
    for i in range(int(n)-1):
        y0=y[i]+diferencial(x[i],y[i])*h
        x[i+1]=x[i]+h
        y[i+1]=y[i]+(((diferencial(x[i],y[i]))+(diferencial(x[i+1],y0)))/(2))*h
        
    return x,y  
```

```python
pasoGraf=0.01
xi=0 #x inicial
yi=1 #y inicial
xf=4 #x final
xexacta=np.arange(xi,xf+pasoGraf,pasoGraf)
yexacta=solucionexacta(xexacta)

h=0.1 #Paso de evolucion inicial.
xexactaErr=np.arange(xi,xf+h,h)
yexactaErr=solucionexacta(xexactaErr)

plt.plot(xexacta,yexacta,label="$Exacta -> y=-x^4/2+4x^3-10x^2+8.5x+1$")
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Solucin Ecuacion diferencial ')
plt.legend()

xHeun,yHeun=Heun(xi,xf,yi,h)
plt.plot(xHeun,yHeun,label="$y Heun$")
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Solucin Mtodo de Heun, $f(x,y)=-2x^3+12x^2-20x+8.5$')
plt.legend()
plt.show()

#Calculo error y error relativo

E= np.abs(yexactaErr-yHeun)
plt.plot(xHeun,E,label="$Error$")
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Error')
plt.show()

ER= np.abs(yexactaErr-yHeun)/(np.abs(yexactaErr))*100

plt.plot(xHeun,ER,label="$Error %$")
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Error relativo %')
```
