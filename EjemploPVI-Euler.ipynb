## Ejemplo Problema de valor inicial

Solución exacta de la ecuación diferencial

$y(x)'=-2x^3+12x^2-20x+8.5$ en el intervalo $x\in[0,4]$ con condición inicial $x=0$ , $y=1$

La solución es:

$y(x)=-0.5x^4+4x^3 -10x^2 +8.5x+C$

Usando la condición inicial, se despeja $C$ y se tiene la solución:

$y=-0.5x^4+4x^3 -10x^2 +8.5x+1$



```python
"""
Simulación
Ingeniería en Multimedia Campus
Universidad Militar Nueva Granada

Código que grafica la solución de la ecuación diferencial
y(x)'=-2x^3+12x^2-20x+8.5 para x en el intervalo [0,4] con condición inicial x=0 , y=1

Elaborado por Eduard Sierra
8 de Agosto de 2022

"""
import matplotlib.pyplot as plt
import numpy as np

#Funcion para evaluara la ecuación de la solción exacta
def SolucionExacta(x):
    return -0.5*x**4+4*x**3 -10*x**2 +8.5*x+1



x=np.linspace(0,4,41)#es de 41 elementos para ir cada 0.1 en el intervalo [0,4]
#print(x)
y=SolucionExacta(x)
#print(y)

plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Solución exacta de $-2x^3+12x^2-20x+8.5$, y(0)=1')
#plt.legend('Exacta $y=-0.5x^4+4x^3 -10x^2 +8.5x+1$')
plt.legend('Exacta')
plt.grid()

# Generamos el grafico.
plt.show()
```

## Solución método de Euler ecuación diferencial

$\frac{dy}{dt}=f(x,y)=-2x^3+12x^2-20x+8.5$ , condición inicial $t=0$, $x=0$, $y=1$

$y_{i+1} = y_i + f(x_i,y_i)h$

```python
import matplotlib.pyplot as plt
import numpy as np

def diferencial(x,y):
    return -2*x**3 +12*x**2 - 20*x + 8.5

def solucionexacta(x):
    return -(x**4)/2 + 4*x**3-10*x**2+8.5*x+1

def Euler(xi,xf,yi,h):
    n=((xf-xi)/h)+1
    x=np.empty(int(n))
    y=np.empty(int(n))
    x[0]=xi
    y[0]=yi
    for i in range(int(n)-1):
        y[i+1]=y[i]+diferencial(x[i],y[i])*h
        x[i+1]=x[i]+h

    return x,y  
```

```python
pasoGraf=0.01
xi=0 #x inicial
yi=1 #y inicial
xf=4 #x final
xexacta=np.arange(xi,xf+pasoGraf,pasoGraf)
yexacta=solucionexacta(xexacta)

h=0.1 #Paso de evolucion inicial.
xexactaErr=np.arange(xi,xf+h,h)
yexactaErr=solucionexacta(xexactaErr)

plt.plot(xexacta,yexacta,label="$Exacta -> y=-x^4/2+4x^3-10x^2+8.5x+1$")
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Solución Ecuacion diferencial ')
plt.legend()

xEuler,yEuler=Euler(xi,xf,yi,h)
plt.plot(xEuler,yEuler,label="$y Euler$")
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Solución Método de Euler, $f(x,y)=-2x^3+12x^2-20x+8.5$')
plt.legend()
plt.show()

#Calculo error y error relativo

E= np.abs(yexactaErr-yEuler)
plt.plot(xEuler,E,label="$Error$")
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Error')
plt.show()

ER= np.abs(yexactaErr-yEuler)/(np.abs(yexactaErr))*100

plt.plot(xEuler,ER,label="$Error %$")
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Error relativo %')

```
