```python
import matplotlib.pyplot as plt
import numpy as np

def diferencial(x,y):
    return -2*x**3 +12*x**2 - 20*x + 8.5

def solucionexacta(x):
    return -(x**4)/2 + 4*x**3-10*x**2+8.5*x+1

def Kutta(xi,xf,yi,h):
    n=((xf-xi)/h)+1
    x=np.empty(int(n))
    y=np.empty(int(n))
    
    k1=0
    k2=0
    
    x[0]=xi
    y[0]=yi
    for i in range(int(n)-1):
        
        k1=diferencial(x[i],y[i])
        k2=diferencial(x[i]+(p1*h),y[i]+(q1*k2*h))
        y[i+1]=y[i]+((a1*k1)+(a2*k2))*h
        x[i+1]=x[i]+h

    return x,y  
```

```python
a2=1/2 #Heun
a1=1-a2
p1=1/(2*(a2))
q1=1/(2*(a2))
    
pasoGraf=0.01
xi=0 #x inicial
yi=1 #y inicial
xf=4 #x final
xexacta=np.arange(xi,xf+pasoGraf,pasoGraf)
yexacta=solucionexacta(xexacta)

h=0.1 #Paso de evolucion inicial.
xexactaErr=np.arange(xi,xf+h,h)
yexactaErr=solucionexacta(xexactaErr)

plt.plot(xexacta,yexacta,label="$Exacta -> y=-x^4/2+4x^3-10x^2+8.5x+1$")
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Solucin Ecuacion diferencial ')
plt.legend()

xKutta,yKutta=Kutta(xi,xf,yi,h)
plt.plot(xKutta,yKutta,label="$y Kutta$")
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Solucin Mtodo de Kutta, $f(x,y)=-2x^3+12x^2-20x+8.5$')
plt.legend()
plt.show()

#Calculo error y error relativo

E= np.abs(yexactaErr-yKutta)
plt.plot(xKutta,E,label="$Error$")
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Error')
plt.show()

ER= np.abs(yexactaErr-yKutta)/(np.abs(yexactaErr))*100

plt.plot(xKutta,ER,label="$Error %$")
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Error relativo %')
```
