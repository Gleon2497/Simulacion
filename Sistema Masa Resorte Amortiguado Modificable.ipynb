Sistema Masa Resorte Amortiguado

```python
import numpy as np
import matplotlib.pyplot as plt

#datos
m = 15 #Masa
c = 1 #Constante de amortiguamiento
k = 2 #Constante de elasticidad

#condiciones iniciales
t = 0 #Tiempo inicial
xinc = 2 #Posicin Inicial
vinc = 0 #Velocidad Inicial
u = np.array([xinc,vinc]) #vector de las condiones iniciales u0 y u1

def F(u,t):
    return np.array([u[1],-c*u[1]/m-k*u[0]/m])
    # return np.array([u[1],m-k*u[0]/m]) Aqui el modelo es ideal
#solucion
tsol = [t] #Agregar condicion inicial al vector del tiempo 
xsol=[u[0]] #Agregar condicion inicial al vector de posicion
vsol=[u[1]] #Agregar condicion inicial al vector de velocidad 
dt = 0.8 #Paso
tfin = 100 #Segundos

#Rugen Kuta
while t<tfin:
    k1=F(u,t)
    k2=F(u+dt*k1/2,t+dt/2)
    k3=F(u+dt*k2/2,t+dt/2)
    k4=F(u+dt*k3,t+dt)
    u = u +(k1+2*k2+2*k3+k4)*dt/6
    t = t+dt 
    xsol.append(u[0])
    vsol.append(u[1])
    tsol.append(t)

plt.plot(tsol,xsol)
plt.show()

#Grafico 3D 
from mpl_toolkits.mplot3d import axes3d
fig=plt.figure()
ax=fig.gca(projection='3d')
ax.plot(xsol,vsol,tsol)
plt.show()
```
